```python
fert_name = {'Urea': 1,'DAP': 2,'14-35-14':3, '28-28':4, '17-17-17':5, '20-20':6,
       '10-26-26':7}
fert['Fertilizer Name'] = [fert_name[item] for item in fert['Fertilizer Name']]
```

```python
fert
```

```python
crop_name = { 'Maize':1, 'Sugarcane':2, 'Cotton':3 ,'Tobacco':4, 'Paddy':5, 'Barley':6,
       'Wheat':7, 'Millets':8, 'Oil seeds':9, 'Pulses':10, 'Ground Nuts':11 }
soil_name ={'Sandy':1, 'Loamy':2, 'Black':3, 'Red':4, 'Clayey':5}
```

```python
fert['Crop Type'] = [crop_name[item] for item in fert['Crop Type']]
fert['Soil Type'] = [soil_name[item] for item in fert['Soil Type']]
```

```python
fert.head()
```

```python
x = fert.drop("Fertilizer Name",axis=1)
y = fert[["Fertilizer Name"]]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33, random_state=42)
print("x_train :",x_train.shape)
print("x_test :",x_test.shape)
print("y_train :",y_train.shape)
print("y_test :",y_test.shape)
```
